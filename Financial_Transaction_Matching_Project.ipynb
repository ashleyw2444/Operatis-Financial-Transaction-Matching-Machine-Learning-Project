# **MATCHING**

```python
# Import the pandas library and alias it as 'pd'
import pandas as pd

# Function to separate transactions based on debit/credit and amount sign
def separate_transactions(transactions, amount):
    transactions_positive = transactions[transactions[amount] >= 0].copy()
    transactions_negative = transactions[transactions[amount] < 0].copy()

    return transactions_positive, transactions_negative

# Separate b_transactions based on 'DR' and 'CR'
b_transactions_dr = b_transactions_sorted[b_transactions_sorted['B_debitOrCredit'] == 'DR'].copy()
b_transactions_cr = b_transactions_sorted[b_transactions_sorted['B_debitOrCredit'] == 'CR'].copy()

# Separate a_transactions based on 'DR' and 'CR'
a_transactions_dr = a_transactions_sorted[a_transactions_sorted['A_debitOrCredit'] == 'DR'].copy()
a_transactions_cr = a_transactions_sorted[a_transactions_sorted['A_debitOrCredit'] == 'CR'].copy()

# Separate transactions for each category and sign for b_transactions
b_transactions_dr_positive, b_transactions_dr_negative = separate_transactions(b_transactions_dr, 'B_amount')
b_transactions_cr_positive, b_transactions_cr_negative = separate_transactions(b_transactions_cr, 'B_amount')

# Merge positive and negative b_transactions with a_transactions
merged_b_transactions_cr_positive = pd.merge(
    a_transactions_cr,
    b_transactions_cr_positive,
    how='inner',
    left_on=['A_valueDate'],
    right_on=['B_valueDate']
)

merged_b_transactions_cr_negative = pd.merge(
    a_transactions_cr,
    b_transactions_cr_negative,
    how='inner',
    left_on=['A_valueDate'],
    right_on=['B_valueDate']
)

merged_b_transactions_dr_positive = pd.merge(
    a_transactions_dr,
    b_transactions_dr_positive,
    how='inner',
    left_on=['A_valueDate'],
    right_on=['B_valueDate']
)

merged_b_transactions_dr_negative = pd.merge(
    a_transactions_dr,
    b_transactions_dr_negative,
    how='inner',
    left_on=['A_valueDate'],
    right_on=['B_valueDate']
)

# Concatenate all matched pairs
matched_df = pd.concat([
    merged_b_transactions_cr_positive,
    merged_b_transactions_cr_negative,
    merged_b_transactions_dr_positive,
    merged_b_transactions_dr_negative
])

# Add match_flag column
matched_df['match_flag'] = (abs(matched_df['A_amount'] - matched_df['B_amount']) <= 2).astype(int)

# Display the matched DataFrame
print(matched_df)

```

```python
# Organize matched_df based on match_flag
matched_df_sorted = matched_df.sort_values(by='match_flag', ascending=False)

# Display the sorted DataFrame
print(matched_df_sorted)
```

# **RANDOM FOREST**

```python
# Import the pandas library and alias it as 'pd'
import pandas as pd

# Import necessary modules from scikit-learn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Assuming you have a DataFrame 'matched_df' containing your data

# Set a seed for reproducibility
seed = 42

# Define the list of features based on column names in the DataFrame
features = [
    'A_amount', 'B_amount'
    # Add any other relevant columns
]

# Define the label column in the DataFrame
labels = 'match_flag'

# Ensure consistent data types by converting specific columns to numeric
matched_df['A_amount'] = pd.to_numeric(matched_df['A_amount'])
matched_df['B_amount'] = pd.to_numeric(matched_df['B_amount'])

# Create a DataFrame with selected features and labels
feature_df = matched_df[features]
label_df = matched_df[labels]

# Split the data into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(feature_df, label_df, test_size=0.7, random_state=seed)

# Initialize the Random Forest Classifier with a specified random seed
rf_model = RandomForestClassifier(random_state=seed)

# Train the model using the training data
rf_model.fit(X_train, y_train)

# Make predictions on the test set
predictions = rf_model.predict(X_test)

# Evaluate the model performance using accuracy, confusion matrix, and classification report
accuracy = accuracy_score(y_test, predictions)
conf_matrix = confusion_matrix(y_test, predictions)
class_report = classification_report(y_test, predictions)

# Display the evaluation metrics
print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Classification Report:\n{class_report}")

```
